```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Definir la función objetivo
x1, x2, lambda1 = sp.symbols("x1 x2 lambda1")
f = x1**2 + x2**2  # Función objetivo

# Restricción
g = x1 + x2 - 1  # Restricción g(x1, x2) = 0

# Crear la función de Lagrange
L = f + lambda1 * g
print(f"Función de Lagrange: L = {L}")

# Derivar L con respecto a x1, x2 y lambda1
dL_dx1 = sp.diff(L, x1)
dL_dx2 = sp.diff(L, x2)
dL_dlambda = sp.diff(L, lambda1)

# Resolver el sistema de ecuaciones
soluciones = sp.solve([dL_dx1, dL_dx2, dL_dlambda], [x1, x2, lambda1], dict=True)
print("\nSoluciones encontradas:")
for sol in soluciones:
    print(sol)

# Evaluar la función objetivo en las soluciones si las hay
if soluciones:
    valores_objetivo = [f.subs(sol) for sol in soluciones]
    print("\nValores de la función objetivo en las soluciones:")
    for i, valor in enumerate(valores_objetivo):
        print(f"Solución {i+1}: f = {valor}")
else:
    print("\nNo se encontraron soluciones para el sistema.")

# Graficar las soluciones
print("\nGenerando el gráfico...")

# Crear un grid para dibujar la función y la restricción
x = np.linspace(-1, 2, 400)
y = np.linspace(-1, 2, 400)
X, Y = np.meshgrid(x, y)
F = X**2 + Y**2
G = X + Y - 1

# Graficar la función objetivo y la restricción
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, F, levels=20, cmap='viridis', alpha=0.8)
plt.clabel(contour, inline=True, fontsize=8, fmt="%.1f")
plt.contour(X, Y, G, levels=[0], colors="red", linewidths=2, label="Restricción")

# Agregar las soluciones encontradas al gráfico
for sol in soluciones:
    plt.scatter(float(sol[x1]), float(sol[x2]), color="blue", label=f"Solución ({float(sol[x1]):.2f}, {float(sol[x2]):.2f})")

# Personalizar el gráfico
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.title("Método de los Multiplicadores de Lagrange")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend(loc="best")
plt.grid(alpha=0.3)
plt.show()

```
